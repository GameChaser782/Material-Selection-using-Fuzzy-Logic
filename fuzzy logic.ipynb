```python
import pandas as pd
from math import log
```

```python
df = pd.read_csv('MM319_data.csv')
values = pd.read_csv('values.csv')
```

```python
Y = []
```

```python
def eqn3(X, X_c, X_max):
    
    if X_c != X_max/2:
        a1 = -100/log(X_c/(X_max - X_c))
        b1 = (X_max - 2*X_c)/[X_c*(X_max - X_c)]
        c1 = X_c/(X_max - X_c)
        return a1*log(b1*X + c1)
    else:
        return 200*(X/X_max) - 100

def eqn4(X, X_c, X_min):
    
    if X_c != 2*X_min:
        a2 = -100/log(-X_min/(X_min - X_c))
        b2 = (-X_c**2 + 2*X_min*X_c)/(X_min - X_c)
        c2 = -X_min/(X_min - X_c)
        return a2*log((b2/X) + c2)
    else:
        return 200*(X_min/X) - 100
```

```python
for i in range(df.shape[0]):
    
    Y1 = eqn3(df.iloc[i][1], values.iloc[0][2], df['Specific Strength (MPa)'].max())
    Y2 = eqn3(df.iloc[i][2], values.iloc[1][2], df['Specific Modulus (GPa)'].max())
    Y3 = eqn3(df.iloc[i][3], values.iloc[2][2], df['Corosion Resistance'].max())
    Y4 = eqn4(df.iloc[i][4], values.iloc[3][2], df['Cost Category'].min())
    
    Y.append(values.iloc[0][1]*Y1 + values.iloc[1][1]*Y2 + values.iloc[2][1]*Y3 + values.iloc[3][1]*Y4)

df['Scaling (Y)'] = Y
```

```python
df['Rank'] = df['Scaling (Y)'].rank(ascending=False)
display(df)
```
